```python
# importing two modules named selenium and pandas
# pandas for converting scraped data into exel file
# selenium for scraping data

import pandas as pd
from selenium import webdriver
```

```python
# run the chrome driver .exe file and then paste link of site from where you want to scrape the data

browser = webdriver.Chrome("chromedriver.exe")
browser.get("https://in.linkedin.com/jobs/search?keywords=Engineering&location=India&geoId=102713980&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0")
```

```python
job_posotion = browser.find_elements_by_class_name("base-search-card__title")
company_name = browser.find_elements_by_class_name("base-search-card__subtitle")
job_location = browser.find_elements_by_class_name("job-search-card__location")


job_posotion_l = []
company_name_l = []
job_location_l = []

for i in job_posotion:
    job_posotion_l.append(i.text)
    
print(job_posotion_l)
print(len(job_posotion_l))


for i in company_name:
    company_name_l.append(i.text)
    
print(company_name_l)
print(len(company_name_l))


for i in job_location:
    job_location_l.append(i.text)
print(job_location_l)
print(len(job_location_l))
```

```python
#creating dataframe from existing extracted categories 

df = pd.DataFrame({'Company': company_name_l, 'Job Position': job_posotion_l, 'Location': job_location_l})
df
```

```python
#importing sqlalchemy

import sqlalchemy

#creatiing engine to connect to database

engine = sqlalchemy.create_engine('mysql+pymysql://root:@localhost:3306/scraped_data')
```

```python
#writing into mysql database from dataframe

df.to_sql(
    name='Jobs', # database table name
    con=engine,
    if_exists='append',
    index=False
)
```

# Saving states names in mysql database

```python
states = browser.find_elements_by_class_name("job-search-card__location")


states_l = []
for i in states:
    states_l.append(i.text)
print(states_l)
print(len(states_l))
```

```python
states_name = []
for i in states_l:
    i = i.split(',')
    states_name.append(str(i[1:2]))
```

```python
#creating dataframe from existing extracted categories 

df = pd.DataFrame({'State': states_name})
df
```

```python
#importing sqlalchemy

import sqlalchemy

#creatiing engine to connect to database

engine = sqlalchemy.create_engine('mysql+pymysql://root:@localhost:3306/scraped_data')
```

```python
#writing into mysql database from dataframe

df.to_sql(
    name='states', # database table name
    con=engine,
    if_exists='append',
    index=False
)
```
